```python
import pandas as pd
set

threatened_species= pd.read_excel('Threatened Species.xlsx')
threatened_species[:10]
```

```python
#Statistics
#summaries
display(threatened_species[["Year", "Value", "Area"]].groupby(["Year", "Area"]).sum())

#descriptives
print(threatened_species.agg(
     {      "Series": ["min", "max", "mean"],
            "Area": ["min", "max", "mean"],
            "Year": ["min", "max", "mean"]
     }
))


print(threatened_species.groupby(["Year", "Series"]).agg(
    {
         'Value':["min","max","mean", "std", "sum"]
            
    }
))
```

```python
#three questions
#1 Has the total number of threatened species increased from 2010 to 2020? 
#2 Which category of species has the largest increase in the last 10 years? 
#3 Which country has the largest increase in the last 10 years? 
```

```python
#Question 1:
start_count=threatened_species[["Year", "Value", "Area"]].groupby(["Year", "Area"]).sum()
subset_10 = threatened_species[threatened_species["Year"] ==2010]
sum_10 = subset_10["Value"].sum()
subset_20 = threatened_species[threatened_species["Year"] ==2020]
sum_20=subset_20["Value"].sum()
Q1_result= sum_20-sum_10
question_1= 'The total number of threatened species has increased by:'
print(question_1,Q1_result)
```

```python
import numpy as np
#Question 2
#subset
series_10=(subset_10.groupby("Series").agg(
    {
         'Value':["sum"]
            
    }
))
series_20=(subset_20.groupby("Series").agg(
    {
         'Value':["sum"]
            
    }
))
difference_v2= series_20-series_10
Q2_result=difference_v2.sort_values(difference_v2.columns[0],ascending=False)
question_2= 'The category with the largest increase in threatened species is:'
row1= Q2_result.head(1)
print(question_2, row1)
```

```python
#Question 3
string= 'Threatened Species: Total (number)'

#2010
country_year = threatened_species[(threatened_species["Year"] ==2010)]
country_total_2010= country_year[(country_year["Series"] == string)]

#2020
country_year20 = threatened_species[(threatened_species["Year"] ==2020)]
country_total_2020= country_year20[(country_year20["Series"] == string)]

display(country_total_2020)

#different
sum_country_2010=country_total_2010[['Value', 'Area']].groupby("Area").sum()
sum_country_2020=country_total_2020[['Value', 'Area']].groupby("Area").sum()

results= sum_country_2020-sum_country_2010
results_v2= results.dropna()
Q3_result=results_v2.sort_values(by="Value", ascending=False)
display(Q3_result[:10])
question_3= 'The country with the largest increase in threatened species is:'
row1= Q3_result.head(1)
print(question_3, row1)
```
